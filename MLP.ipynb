## Walkthrough of MLP code

This is a full walk through the code for a multilayer perceptron, implemented in Python2. 

A multilayer perceptron (MLP) is a feedforward neural network which generalizes the standard perceptron by having one or more hidden layers. By the universal approximation theorem [1], an MLP with one hidden layer containing a finite number of neurons can approximate any continuous function. Therefore, this implementation is restricted to the 'classical' MLP with one input layer, one hidden layer, and one output layer.
